```python
def sort_sqft(i):
    tempval = i.split('-')
    if len(tempval) >= 2:
        tempval = [float(x) for x in tempval]
        return sum(tempval)/len(tempval)
    try:
        return float(i)   
    except:
        return None       
```

```python
df['total_sqft']  = df['total_sqft'].apply(sort_sqft)
```

```python
## price is in lakhs so we need price per square ft
df['price_per_sqft'] = df['price'] * 100000/df['total_sqft']
```

## outlier detection and removal 

```python
df.describe()
```

```python
def remove_outliers(df):
    remove_outlier_df = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        mean = np.mean(subdf.price_per_sqft)
        stddev = np.std(subdf.price_per_sqft)
        new_df = subdf[ ((subdf.price_per_sqft > (mean-stddev)) & (subdf.price_per_sqft <= (mean+stddev)) )]
        remove_outlier_df = pd.concat([remove_outlier_df,new_df], ignore_index = False)
    return remove_outlier_df        
```

```python
df = remove_outliers(df)
```

```python
df.shape
```

```python
df.describe()
```

```python
def remove_outliers_bhk(df):
    exculde_values = np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] ={
                "mean" : np.mean(bhk_df.price_per_sqft),
                "std": np.std(bhk_df.price_per_sqft),
                "count":bhk_df.shape[0]
            }
        print(location,bhk_stats)
        print('*'*100)
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats["count"]> 5:
                exculde_values = np.append(exculde_values,bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
    print(exculde_values)
    print(exculde_values.shape)
    return df.drop(exculde_values,axis='index')
```

```python
df = remove_outliers_bhk(df)
```

# Training DataSet

```python
## importing libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score,r2_score
```

```python
X = df.drop(columns = ['price'])
y = df['price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state= 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Apply Linear Regression


```python
column_transform = make_column_transformer((OneHotEncoder(sparse=False),['location']),remainder='passthrough')
```

```python
scalar = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_transform,scalar,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

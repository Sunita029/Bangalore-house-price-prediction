## Fixing Outliers

```python
df['total_sqft'].unique()
```

```python
def sort_sqft(i):
    tempval = i.split('-')
    if len(tempval) >= 2:
        tempval = [float(x) for x in tempval]
        return sum(tempval)/len(tempval)
    try:
        return float(i)   
    except:
        return None       
```

```python
df['total_sqft']  = df['total_sqft'].apply(sort_sqft)
```

```python
df.sample(20)
```

```python
## price is in lakhs so we need price per square ft
df['price_per_sqft'] = df['price'] * 100000/df['total_sqft']
```

```python
df.sample(20)
```

```python
df['location'].value_counts()  
## we can't do OHE wo se have to reduce the 
#locations so we replce the values of location having < 10 named as others
```

```python
df['location'] = df['location'].apply(lambda x : x.strip())

## removing blank spaces from location text
```

```python
df.sample(20)
```

```python
location_counts = df['location'].value_counts()  
```

```python
location_counts_less_than_10 = location_counts[location_counts<=10]
```

```python
df['location'] = df['location'].apply(lambda x: "others" if x in location_counts_less_than_10 else x )
```

## outlier detection and removal 

```python
df.describe()
```

```python
df = df[((df['total_sqft']/df['BHK'])>= 300 )]
```

```python
df['price_per_sqft'].describe()
```

```python
def remove_outliers(df):
    remove_outlier_df = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        mean = np.mean(subdf.price_per_sqft)
        stddev = np.std(subdf.price_per_sqft)
        new_df = subdf[ ((subdf.price_per_sqft > (mean-stddev)) & (subdf.price_per_sqft <= (mean+stddev)) )]
        remove_outlier_df = pd.concat([remove_outlier_df,new_df], ignore_index = False)
    return remove_outlier_df        
```

```python
df = remove_outliers(df)
```
